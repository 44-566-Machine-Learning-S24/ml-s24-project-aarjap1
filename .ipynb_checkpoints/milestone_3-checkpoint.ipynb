```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('StudentsPerformance.csv')

mapping = {
    "bachelor's degree": 1,
    'some college': 2,
    "master's degree": 3,
    "associate's degree": 4
}

mapping2 = {
    "male": 1,
    "female": 2
}

mapping3 = {
    "group A": 1,
    "group B": 2,
    "group C": 3,
    "group D": 4,
    "group E": 5
}

data['encoded_education_level'] = data['parental level of education'].map(mapping)
data['encoded_gender'] = data['gender'].map(mapping2)
data['encoded_race'] = data['race/ethnicity'].map(mapping3)

data["Total_Score"] = data["math score"] + data["reading score"] + data["writing score"]

data = data.dropna()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
warnings.filterwarnings("ignore", category=UserWarning)

# Clustering
X_cluster = data[['Total_Score', 'encoded_education_level', 'encoded_gender', 'encoded_race']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
data['cluster'] = kmeans.fit_predict(X_scaled)

# Dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
data['PCA1'] = X_pca[:, 0]
data['PCA2'] = X_pca[:, 1]

plt.figure(figsize=(12, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='cluster', data=data, palette='viridis', alpha=0.7)
plt.title('PCA Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

Explanation:
I chose 4 clusters because there are 4 different parental levels I am looking to classify into. Looking at the plot above two of clusters are squished together with each other. The purple clusters occupy the right hand-side and the yellow occupy top left and middle side. The green occupy the bottom left and is also squished in the middle with the blue clusters.

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest Classifier
X_rf = data[['Total_Score', 'encoded_gender', 'encoded_race']]
y_rf = data['encoded_education_level']

rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_rf, y_rf)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix

rf_cv_scores = cross_val_score(rf_classifier, X_rf, y_rf, cv=5)
print("Random Forest Classifier Cross-Validation Scores:", rf_cv_scores)
print("Mean CV Score:", rf_cv_scores.mean())

print("\nRandom Forest Classifier Classification Report:")
rf_y_pred = rf_classifier.predict(X_rf)
print(classification_report(y_rf, rf_y_pred))
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.heatmap(confusion_matrix(y_rf, rf_y_pred), annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
```

Explanation on Random forest:

The weighted average for the all the performance metrics(precision, recall, f1-score) we got was 87%. The model did overall well looking at the confusion matrix and the performance metrics.
    

```python
from sklearn.neural_network import MLPClassifier
#splitting test and training set
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)

# Neural Network Classifier
X_nn = train_set[['Total_Score', 'encoded_gender', 'encoded_race', 'encoded_education_level']]
y_nn = train_set['encoded_education_level']

X_nn_test = test_set[['Total_Score', 'encoded_gender', 'encoded_race', 'encoded_education_level']]
y_nn_test = test_set['encoded_education_level']

nn_classifier = MLPClassifier(random_state=42, max_iter=1000)
nn_classifier.fit(X_nn, y_nn)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix

# Neural Network Classifier Evaluation
nn_cv_scores = cross_val_score(nn_classifier, X_nn_test, y_nn_test, cv=5)
print("Neural Network Classifier Cross-Validation Scores:", nn_cv_scores)
print("Mean CV Score:", nn_cv_scores.mean())

print("\nNeural Network Classifier Classification Report:")
nn_y_pred = nn_classifier.predict(X_nn_test)
print(classification_report(y_nn_test, nn_y_pred))
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
sns.heatmap(confusion_matrix(y_nn_test, nn_y_pred), annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Neural Network Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')

plt.tight_layout()
plt.show()
```
